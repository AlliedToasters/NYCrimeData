This project uses the New York State crime statistics data for 2013, taken from <a href='https://ucr.fbi.gov/crime-in-the-u.s/2013/crime-in-the-u.s.-2013/tables/table-8/table-8-state-cuts/table_8_offenses_known_to_law_enforcement_new_york_by_city_2013.xls'>the FBI UCR repository</a>.<br><br>

In this data set, "Property Crime" is defined as the sum of reported burglary, larceny-theft, and motor vehicle theft. The assignment is to build a linear regression model to predict property crime, so I'll build features out of the other field.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline
#df[['cbrt_EPCpc', 'Murd', 'Violent crime_per_capita']] WORKS GREAT!
```

```python
df = pd.read_excel('clean_crime_data.xlsx')
cp = df.copy()
cp.head()
```

Before engineering most features, I transform them to per capita crime rates.
<br><br>
First, following some handy advice from the <a href'https://statswithcats.wordpress.com/2010/11/21/fifty-ways-to-fix-your-data/'> stats with cats blog</a>, I apply a cube root transform to the target variable (Property crime per capita) to make it look more normal.

```python
df['PC_per_capita'], df['Burg_per_capita'] = df['Property crime']/df.Population, df['Burglary']/df.Population
df['LT_per_capita'], df['MVT_per_capita'] = df['Larceny-theft']/df.Population, df['Motor vehicle theft']/df.Population
df['cbrt_PCpc'] = np.cbrt(df['PC_per_capita'])
df['has_LT'] = np.where((df['Larceny-theft'] > 0), 1, 0)
df['has_Burg'] = np.where((df['Burglary']), 1, 0)
df['has_MVT'] = np.where((df['Motor vehicle theft']), 1, 0)
```

That looks kind of normal! Now I'll build my first input feature, called 'Murd', which is a binary feature describing if any murders were reported.

```python
df['cbrt_Bpc'] = np.cbrt(df['Burg_per_capita'])
df['cbrt_LTpc'] = np.cbrt(df['LT_per_capita'])
df['cbrt_MVTpc'] = np.cbrt(df['MVT_per_capita'])

plt.hist(df['cbrt_Bpc'], alpha=.5, label='Burglary');
plt.hist(df['cbrt_LTpc'], alpha=.5, label='Larceny-theft');
plt.hist(df['cbrt_MVTpc'], alpha=.5, label='Motor vehicle theft');
plt.legend();
```

```python
plt.hist(np.cbrt(df[df['has_Burg'] == 1]['cbrt_Bpc']), alpha=.5, label='Burglary');
plt.hist(np.cbrt(df[df['has_LT'] == 1]['cbrt_LTpc']), alpha=.5, label='Larceny-theft');
plt.hist(np.cbrt(df[df['has_MVT'] == 1]['cbrt_MVTpc']), alpha=.5, label='Motor vehicle theft');
plt.legend();
```

```python
g = sns.PairGrid(df[['cbrt_Bpc', 'cbrt_LTpc', 'cbrt_MVTpc']], diag_sharey=False)
g.map_upper(plt.scatter, alpha=.5)
g.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
g.map_diag(sns.kdeplot, lw=3)
plt.show()
```

```python
df['Murd_pc'] = df['Murder and nonnegligent manslaughter']/df.Population
df['Rape_pc'] = df['Rape (legacy definition)2']/df.Population
df['Aslt_pc'] = df['Aggravated assault']/df.Population
df['4thrt_AApc'] = np.power(df['Aslt_pc'], (1/4))
df['VCpc'] = df['Violent crime']/df.Population
df['4thrt_VCpc'] = np.power(df['VCpc'], (1/4))
df['has_VC'] = np.where((df['Violent crime'] > 0), 1, 0)
df['has_Rape'] = np.where((df['Rape (revised definition)1'] > 0), 1, 0)
df['has_Rob'] = np.where((df['Robbery'] > 0), 1, 0)
df['has_Murd'] = np.where((df['Murder and nonnegligent manslaughter'] > 0), 1, 0)
df['no_VC'] = np.where((df['Violent crime'] == 0), 1, 0)
```

```python
df['known_arson'] = np.where((df['Arson3'].notnull()), df['Arson3'], 0)
df['has_arson'] = np.where((df['Arson3'] > 0), 1, 0)
ars = df[df['has_arson'] == 1]
ars_regr = linear_model.LinearRegression()
murds = ars['Murder and nonnegligent manslaughter'].values.reshape(-1, 1)
arsn = ars.known_arson.values.reshape(-1, 1)
ars_regr.fit(murds, arsn)
ars_regr.score(murds, arsn)
df['pred_arson'] = ars_regr.predict(df['Burglary'].values.reshape(-1, 1))
df['pred_arson'] = np.where((df['pred_arson'] < 0), 0, 1)
df['est_arson'] = np.where((df['has_arson'] == 1), df['Arson3'], df['pred_arson'])
df['est_arson_per_capita'] = df['est_arson']/df.Population
df['4thrt_ars_pc'] = np.power((df['est_arson_per_capita']), (1/4))
```

```python
sns.jointplot(x=df['4thrt_ars_pc'], y=df['4thrt_VCpc']);
```

```python
sns.jointplot(x='cbrt_PCpc', y='4thrt_VCpc', data=df);
```

The next feature takes advantage of the fact that "property crimes" are the sum of a number of other types of crimes. I add all of those types of crimes together. One problem is that the data set includes inconsistent data on arson; many cities have null fields for this category, but the metadata indicates this does not mean they have zero arson. I found a strong linear correlation between burglary and arson, so I built a quick-and-dirty linear regression to fill in the gaps in our arson data, and then add all of the relevant property crimes together, finally dividing by population, for a per-capita property crime estimate. Of course, I need to take the cube root to match the transform I did on the outcome variable

```python
from math import log
df['log_pop'] = df['Population'].apply(lambda x: log(x, 1000))
plt.hist(np.log(df['log_pop']));
df['log_pop'].describe()
```

```python
sns.jointplot(x='log_pop', y='4thrt_VCpc', data=df);
```

The last helpful feature is simply violent crime per capita. This takes into account all of the non-property crimes. This feature helps the most without any transformations.

NYC seems to be an important exception, with much less arson per capita than most other places. Let's make a feature just for being NYC.

```python
df['is_NYC'] = np.where((df.City == 'New York'), 1, 0)
df['is_SFT'] = np.where((df.City == 'Seneca Falls Town'), 1, 0)
```

```python
#What the heck. Let's make a model.
from sklearn import linear_model
regr = linear_model.LinearRegression()
#train = df[['4thrt_VCpc', '4thrt_ars_pc', 'is_NYC']]
train = df[['4thrt_VCpc', 'is_NYC', 'no_VC', 'has_Murd', '4thrt_ars_pc']]
target = df[['cbrt_PCpc']]
regr.fit(train, target);
print('\nR-squared:')
print(regr.score(train, target))
df['raw_prediction'] = regr.predict(train)
predictions = df['raw_prediction']
df['pred_PCpc'] = np.power(predictions, 3)
df['pred_PC'] = df['pred_PCpc'] * df.Population
```

```python
df['act_residual'] = df['cbrt_PCpc'] - df['raw_prediction']
plt.hist(df['act_residual']);
```

```python
plt.scatter(y=df['act_residual'], x=df['raw_prediction']);
print(len(df))
```

```python
pred = df['pred_PC']
targ = df['Property crime']
df['residual'] = pred - targ
```

```python
plt.hist(df['residual']);
```

```python
plt.scatter(x=pred, y=df['residual']);
```

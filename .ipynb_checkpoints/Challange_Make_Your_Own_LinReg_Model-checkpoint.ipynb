```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_excel('clean_crime_data.xlsx')
```

```python
df['pop2'] = df.Population*df.Population
df['Rob'] = np.where((df.Robbery > 0), 1, 0)
df['gen_theft'] = np.where((np.logical_or((np.logical_or(df.Robbery > 0, df.Burglary > 0)), (np.logical_or(df['Larceny-theft'] > 0, df['Motor vehicle theft'] > 0)))), 1, 0)
df['theft_per_capita'] = (df.Robbery + df.Burglary + df['Larceny-theft'] + df['Motor vehicle theft'])/df.Population
df['Murd'] = np.where((df['Murder and nonnegligent manslaughter'] > 0), 1, 0)
df['known_arson'] = np.where((df['Arson3'].notnull()), df['Arson3'], 0)
df['has_arson'] = np.where((df['Arson3'] > 0), 1, 0)
```

```python
ars = df[df['has_arson'] == 1]
sns.jointplot(x='known_arson', y='Burglary', data=ars);
```

```python
ars_regr = linear_model.LinearRegression()
burgs = ars.Burglary.values.reshape(-1, 1)
arsn = ars.known_arson.values.reshape(-1, 1)
ars_regr.fit(burgs, arsn)
ars_regr.score(burgs, arsn)
df['pred_arson'] = ars_regr.predict(df['Burglary'].values.reshape(-1, 1))
df['pred_arson'] = np.where((df['pred_arson'] < 0), 0, 1)
df['est_arson'] = np.where((df['has_arson'] == 1), df['Arson3'], df['pred_arson'])
df['est_arson_per_capita'] = df['est_arson']/df.Population
df.columns
```

```python
df['est_PC_per_capita'] = (df['Burglary'] + df['Larceny-theft'] + df['Robbery'] + df['Motor vehicle theft'] + df['est_arson'])/df.Population
df['cbrt_EPCpc'] = np.cbrt(df['est_PC_per_capita'])
```

```python
keys = list(df.columns[2:4]) + list(df.columns[5:12])
for key in keys:
    df[str(key + '_per_capita')] = df[key]/df['Population']
keys_pc = []
for key in keys:
    keys_pc.append(key+'_per_capita')
```

```python
for key in keys_pc:
    plt.hist(np.sqrt(df[key]), alpha = .5, label=key); #square root helps keep distributions looking nice
plt.legend();
```

```python
df['cbrt_rape_pc'] = np.cbrt(df['Rape (legacy definition)2_per_capita'])
```

Looks like everything is more or less ok. The 'Arson3' column has lots of NaNs because of the FBI's decision not to report incomplete arson data, and the Rape(revised definition)1 column is completely NaNs, but we can work around these issues if necessary.<br><br>
<H2>Challenge: Build a Regression Model to Predict Property Crimes</H2><br><br>

The goal of this challenge is to use a linear regression model to predict property crime. The focus is on predictive power, not descriptive power, so I'll take my liberty on transforming the data for the sake of getting better predictions.<br><br>
First, let's take a look at our target variable, Property crime. I'll work with per capita crime rates here, since that seems to make more sense on multiple levels.


```python
plt.hist(df['Property crime_per_capita']);
```

It's not normal at all. Let's take the cube root:

```python
df['cbrt_PC'] = np.cbrt(df['Property crime_per_capita'])
plt.hist(df['cbrt_PC']);
```

That looks much better. Now let's look for correlations between other potential features:

```python
sns.heatmap(df[(keys_pc + ['Population', 'pop2', 'Rob', 'Murd', 'gen_theft'])].corr());
```

It's probably not surprising to see a strong correlation between all types of crimes all across the board. Still, some are more strongly correlated than others. It's important to note that some of these appear to be aggregate features already; violent crimes, for example, is the sum of rape, aggravated assault, and murder. Let's look at the aggregated violent crimes as a feature:


```python
df['fourthrt_VC'] = np.power(df['Violent crime_per_capita'], .25) #It turns out the fourth root looks a lot more normal
plt.hist(df['fourthrt_VC']); #except for all the zeros. We'll roll with it for now.
```

Let's look for more strongly correlated features

```python
sns.heatmap(df[['fourthrt_VC', 'pop2', 'Population', 'Murd', 'gen_theft']].corr());
```

```python
#There's a correlation between the fourthrt_VC and gen_theft, Murd, which makes sense.
#Let's reject Murd for now, since it contains some of the same info as fourthrt_VC.
#Let's look at some more of our continuous variable distributions.
df['log*3_pop'] = np.log(np.log((np.log(df.Population)))) #taking the log three times makes pop look normal-ish.
plt.hist(df['log*3_pop']);
```

```python
sns.jointplot(x='cbrt_PC', y='fourthrt_VC', data=df); #this looks kinda linear
df['has_VC'] = np.where((df['Violent crime'] > 0), 1, 0)
```

```python
sns.jointplot(x='cbrt_PC', y='log*3_pop', data=df); #This looks very roughly linear
```

```python
df['cbrt_theft_per_capita'] = np.cbrt(df['theft_per_capita']) #Since most property crimes are theft, match transform
plt.hist(df['cbrt_theft_per_capita']); #Sure enough, makes this look normal too.
```

```python
sns.jointplot(x='cbrt_PC', y='cbrt_theft_per_capita', data=df); #Aaand that's pretty much as linear as they come.
```

```python
#What the heck. Let's make a model.
from sklearn import linear_model
regr = linear_model.LinearRegression()
train = df[['cbrt_EPCpc', 'Violent crime_per_capita', 'Murd']]
target = df['cbrt_PC']
regr.fit(train, target);
print('\nR-squared:')
print(regr.score(train, target))
```

It's seems pretty obvious that this model would be able to fit so well, since the vast majority of 'property crimes' are theft. Just adding up the thefts would get us really close.<br><br> How well does our model explain the noise?

```python
predictions = regr.predict(df[['cbrt_EPCpc', 'Violent crime_per_capita', 'Murd']])
pred = np.power(predictions, 3)
targ = df['Property crime_per_capita']
```

```python
residual = targ - pred
```

```python
plt.hist(residual); #Residual looks pretty skewed here.
```

```python
plt.scatter(x=pred, y=residual);
```

```python
df['pred'] = pred
```
